## Dataset Description

The dataset contains historical car auction sales prices, scraped from the outside internet sources and collected in 2015.

Each row represents a sale at a used car auction. Each column is described below.

| Column       | Description                                                                                          |
| ------------ | ---------------------------------------------------------------------------------------------------- |
| year         | Car's year of production                                                                             |
| make         | Car make/brand                                                                                       |
| model        | Car model                                                                                            |
| trim         | Car's trim or grade, which specifies a car model's level of performance, features, options or extras |
| body         | Car body type (e.g. Sedan, SUV, Coupe, etc.)                                                         |
| transmission | Car transmission type. The mechanism that moves the power from the engine to the wheels.             |
| vin          | Vehicle Identification Number, a unique code to identify individual motor vehicles.                  |
| state        | The state in which the car was auctioned.                                                            |
| condition    | The condition of the cars being at the time of auction, 5 being the best score and 1 being the worst.|
| odometer     | The distance traveled by a vehicle since being manufactured, in miles                                |
| color        | Exterior color of the car                                                                            |
| interior     | Interior color of the car                                                                            |
| seller       | The seller of the car or car dealer                                                                  |
| mmr          | Manheim\* Market Report, the market estimated price of the cars. An indicator of wholesale prices.   |
| sellingprice | The price a car was sold at in auction.                                                              |
| saledate     | The date on which the car has been sold.                                                             |

\*Manheim is a company established in 1945 as a car auction company. It has grown to a very reputable company in the industry and its MMR estimate is a baseline tool for wholesale car price determination. They base their price calculations on over 10 million transaction over the past 13-month period.

## Data Pre-Processing & Cleaning

_Apply any data preprocessing and/or feature engineering below. Show/output the changes to the dataset._

```python
#Removed '$'sign from mmr and selligprice coloumn in order to change data type to float, then remove 'mi' from odometer to change data type to float
cars_auction['odometer'] = cars_auction['odometer'].str.extract('([0.0-9.9]+)', expand=False).str.strip()
cars_auction['mmr'] = cars_auction['mmr'].str.extract('([0.0-9.9]+)', expand=False).str.strip()
cars_auction['sellingprice'] = cars_auction['sellingprice'].str.extract('([0.0-9.9]+)', expand=False).str.strip()
cars_auction['mmr']=cars_auction['mmr'].astype(float)
cars_auction['odometer']=cars_auction['odometer'].astype(float)
cars_auction['sellingprice']=cars_auction['sellingprice'].astype(float)
cars_auction.info()
```

```python
#removed all hyphens to clean data
cars_auction['color'] = cars_auction['color'].str.replace('—','black')
cars_auction['interior'] = cars_auction['interior'].str.replace('—','black')
cars_auction.tail(7)
```

```python
#renamed the make column to brand to be easily understood as well as renamed odometer,mmr and selligprice to include thier units in the dataframe
#& capitlized the states initials both steps for better  representation
cars_auction.rename(columns={'make': 'brand'}, inplace=True)
cars_auction.rename(columns={'odometer': 'odometer in mi'}, inplace=True)
cars_auction.rename(columns={'mmr': 'mmr in $'}, inplace=True)
cars_auction.rename(columns={'sellingprice': 'sellingprice in $'}, inplace=True)
cars_auction['state']= cars_auction['state'].str.capitalize()
cars_auction.tail(7)
```

```python
#Replaced each car body type by only one consistent name for each one body type , to create a namin convetion that allows effective data grouping
cars_auction=cars_auction.replace({'body':{'sedan':'Sedan','G Sedan':'Sedan','g sedan':'Sedan'}})
cars_auction=cars_auction.replace({'body':{'suv':'SUV'}})
cars_auction=cars_auction.replace({'body':{'E-Series Van':'Minivan','van':'Minivan','Van':'Minivan', 'minivan':'Minivan', 'e-series van':'Minivan',
                                           'promaster cargo van':'Minivan','Transit Van':'Minivan', 'Promaster Cargo Van':'Minivan'}})

cars_auction=cars_auction.replace({'body':{'coupe':'Coupe', 'G Coupe':'Coupe', 'g coupe':'Coupe', 'Genesis Coupe':'Coupe','genesis coupe':'Coupe',
                                           'CTS-V Coupe':'Coupe', 'Q60 Coupe':'Coupe', 'cts coupe':'Coupe', 'Koup':'Coupe', 'Elantra Coupe':'Coupe',
                                           'CTS Coupe':'Coupe'}})

cars_auction=cars_auction.replace({'body':{'convertible':'Convertible','G Convertible':'Convertible','Beetle Convertible':'Convertible',
                                           'G37 Convertible':'Convertible','Q60 Convertible':'Convertible', 'GranTurismo Convertible':'Convertible'
                                           ,'beetle convertible':'Convertible','g convertible':'Convertible','granturismo convertible':'Convertible'}})

cars_auction=cars_auction.replace({'body':{'hatchback':'Hatchback'}})

cars_auction=cars_auction.replace({'body':{'wagon':'Wagon', 'CTS Wagon':'Wagon', 'TSX Sport Wagon':'Wagon', 'tsx sport wagon':'Wagon'}})

cars_auction=cars_auction.replace({'body':{'Crew Cab':'Pickup Truck','Regular Cab':'Pickup Truck','Double Cab':'Pickup Truck', 'Extended Cab':'Pickup Truck',
                                  'SuperCab':'Pickup Truck', 'king cab':'Pickup Truck','King Cab':'Pickup Truck', 'extended cab':'Pickup Truck',
                                  'quad cab':'Pickup Truck', 'supercab':'Pickup Truck', 'Quad Cab':'Pickup Truck', 'crew cab':'Pickup Truck', 'supercrew':'Pickup Truck',
                                  'CrewMax Cab':'Pickup Truck', 'double cab':'Pickup Truck', 'Club Cab':'Pickup Truck', 'Access Cab':'Pickup Truck',
                                  'Cab Plus 4':'Pickup Truck', 'crewmax cab':'Pickup Truck', 'access cab':'Pickup Truck', 'Mega Cab':'Pickup Truck',
                                  'regular cab':'Pickup Truck', 'mega cab':'Pickup Truck', 'regular-cab':'Pickup Truck', 'Xtracab':'Pickup Truck',
                                  'xtracab':'Pickup Truck','SuperCrew':'Pickup Truck'}})
cars_auction
cars_auction['body'].unique()
```

## Data Preparation for Modelling

_Apply any additional data preparation steps needed before modelling below. Show/output the changes to the dataset._

```python
cars_auction.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cars_auction["model"] = le.fit_transform(cars_auction["model"])
len(cars_auction["model"].unique())
```


# **Reason for label encoding:**
We did label encoding to the 'model' variable as it has many number of categories, so label encoding would be more efiicient than one hot encoding.

```python
categorical_columns = ["brand", "body", "transmission", "color"]

cars_auction_encode = pd.get_dummies(data=cars_auction,
                           columns=categorical_columns,
                           drop_first=True,
                           dtype='int8')
```

# **Reason for one hot encoding:**

Those variables have low number of categories, so applying one hot encoding would be more efficient and increase the model's accuracy.

## **Why we chose and normalized those variables:**

We chose those variable as logically and according to data inspection they are correlated with the selling price.

```python
print('\nColumns in data frame after encoding dummy variable:\n',cars_auction_encode.columns.values)
```

```python
numerical_columns = ['odometer in mi', 'condition', 'year']


cars_auction_encode[numerical_columns] = (

    cars_auction_encode[numerical_columns] - cars_auction_encode[numerical_columns].min()

) /(

    cars_auction_encode[numerical_columns].max() - cars_auction_encode[numerical_columns].min()
)


cars_auction_encode.head()
```

## **Why we chose and normalized those variables:**

We chose those variable as logically and according to data inspection they are correlated with the selling price.

As wanted to include those variables in the model so we had to reduce their range by normalizing them.

```python
cars_auction.corr(numeric_only=True)
```

```python
cars_auction_encode.corr(numeric_only=True)
```

```python
from sklearn.model_selection import train_test_split

x = cars_auction_encode.drop(columns=['trim', 'vin', 'interior', 'seller', 'saledate', 'state', 'sellingprice in $', 'mmr in $'], axis=1) # Unused columns

y = cars_auction_encode['sellingprice in $'] # Dependent variable

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=23)

print('Training dataset size:', x_train.shape[0])
print('Testing dataset size:', x_test.shape[0])
```

### **Why we dropped those columns:**
We dropped those columns('vin', 'seller','saledate')since logically the had no correlation with the selling price as well as correlation was low. As for the other columns ('trim', 'interior', 'state') eventhough they logically affect the selling price, their correlation was low compared to other independent variables.

The mmr however had the highest correlation with the selling price; but we dropped it since their correlation is extremely high which results in overfitting.

We also dropped the selling price as it's the dependent variable, as including it as independent variable is not logically right.


## Modelling

_Apply the linear regression model below._

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()

lin_reg.fit(x_train, y_train);
```

## Model Evaluation

_Evaluate the model you applied._

```python
# sklearn regression module
y_pred = lin_reg.predict(x_test)

# Evaluation
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_pred, y_test)
rmse = np.sqrt(mse)
# R_square
R_square = lin_reg.score(x_test,y_test)
print('The root mean square error (RMSE) is: ', rmse)
print('R^2 obtained is:', R_square)
```

# Handling Outliers


```python
sns.boxplot(x=cars_auction_encode["sellingprice in $"])
```

```python
q = cars_auction_encode["sellingprice in $"].quantile(0.99)
q
```

```python
original_length = len(cars_auction_encode)
print(f'{len(cars_auction_encode[cars_auction_encode["sellingprice in $"] > q])} rows removed')

cars_auction_encode = cars_auction_encode[cars_auction_encode["sellingprice in $"] < q]
cars_auction_encode.reset_index(drop = True, inplace = True)
```

```python
from sklearn.model_selection import train_test_split

x = cars_auction_encode.drop(columns=['trim', 'vin', 'interior', 'seller', 'saledate', 'state', 'sellingprice in $', 'mmr in $'], axis=1)  # Unused columns

y = cars_auction_encode['sellingprice in $'] # Dependent variable

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=23)

print('Training dataset size:', x_train.shape[0])
print('Testing dataset size:', x_test.shape[0])
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()

lin_reg.fit(x_train, y_train);
```

```python
# sklearn regression module
y_pred = lin_reg.predict(x_test)

# Evaluation
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_pred, y_test)
rmse = np.sqrt(mse)

# R_square
R_square = lin_reg.score(x_test,y_test)
print('The root mean square error (RMSE) is: ', rmse)
print('R^2 obtained is:', R_square)
```
